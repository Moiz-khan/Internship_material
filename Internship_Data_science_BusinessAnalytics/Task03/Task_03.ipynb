### Removing the outliers in our dataset

```python
#first we need to find out the Inter-Quartile Range(IQR)
Q1 = superStore_dataset.quantile(0.25)
Q3 = superStore_dataset.quantile(0.75)

IQR = Q3-Q1
print(IQR)
```

Here is the Inter-Quartile Range(IQR) of all columns except objects

```python
#Data type error to show oulier values
#printing the outlier values data 
#print(superStore_dataset < (Q1 -1.5 * IQR)) | (superStore_dataset > (Q3 + 1.5 *IQR))
```

```python
superStore_dataset = superStore_dataset[~((superStore_dataset < (Q1 -1.5 * IQR)) | (superStore_dataset > (Q3 + 1.5 *IQR))).any(axis=1)]
superStore_dataset.shape
```

```python
superStore_dataset.boxplot()
```

All outliers that are exist in our Sales, Quantity, discount and profit data has been removed.
Now, there is no outliers in our dataset. Data is ready!!!
